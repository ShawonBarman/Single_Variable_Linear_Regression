```python
# dataset

import os
import io
import pandas as pd

df = pd.read_csv('DATA.csv') 
x = df["x"]
print(x)
y = df["y"]
print(y)
```

```python
# Parameter Initialization

theta = [1, 2]
print(theta[0])
print(theta[1])
```

```python
count = 0

for i in range(200):
    
    print("Iteration number: ", count+1)
    count = count + 1
    print("\n")

    # Hypothesis Function

    m = len(x)
    #print(m)
    h = []

    for i in range(m):
        h.append(theta[0]+theta[1]*x[i])
    print("Hypothesis Function is: ", h) 


    # Cost Function

    error = 0

    for i in range(m):
        error = error + (h[i] - y[i]) ** 2

    j = (1/(2*m)) * error 
    print("Cost Function is: ", j) 


    # Gradient Descent

    sum0 = 0
    sum1 = 0

    for i in range(m):
        sum0 = sum0 + (h[i] - y[i])
        sum1 = sum1 + (h[i] - y[i]) * x[i]

    alpha = 0.01
    theta[0] = theta[0] - alpha/m * sum0
    theta[1] = theta[1] - alpha/m * sum1

    print("Updated Parameters are: ", theta) 


import matplotlib.pyplot as plt
plt.plot(x, y, marker = 'x')
plt.plot(x, h, marker = 'o')

plt.show()
    
    
```
